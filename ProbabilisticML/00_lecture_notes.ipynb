An inference probelm requires a statement about the value of an unobserved (latent) variable $x$ based on observations $y$ which are related to $x$, ut may not be suffcient to fully determine $x$. This requires the notion of **uncertainty**

In daily life, nothing is ever certain, and all information is incomplete. Contrast to theoretical computer science, in machine learning we thus only seek to find good hypotheses, mak prediction, and estimates. This process of inference can be made formal, by anchoring it to classic propositional logic.

## Limitations of propositional logic (deductive and plausible reasoning)

Let $A$ be the statement "It rains", and $B$ be the statement "The street is wet"

### Deductive reasoning
$A\implies B$: If it rains, then the street is wet. But does $B\implies A$? If you look outside, and you find that the street is wet, does that implied it rained? Well, no, but it does _seem plausible_ that it rained! Propositional logic does not allow us to make this kind of reasoning.

### Plausible reasoning
Indeed, $P(B|A) > P(B)$, if $A$ is true, then $B$ becomes more plausible, in particular, plausible reasoning allows us to make statements as

- $A$ is true, thus $B$ becomes more plausible
- $B$ is true, thus $A$ becomes less plausible
- $B$ is true, thus $A$ becomes more plausible
- $A$ is false, thus $B$ becomes less plausible


## The red card problem

There are three cards: the first card is red on both sides (call it $C_{RR}$), the second card is red on one side and white on the other (call it $C_{RW}$), and the last card is white on both sides ($C_{WW}$), so that we have 
$$\mathcal{C} = \{C_{RR}, C_{RW}, C_{WW}\}$$

Now, the three cards are drawn in the bag, which is shaken, then a card is drawn randomly from the bag, and this card is red on one side, what is the probability that it is card $C_{RR}$?

Let $R$ denote the event of seeing red side of the card. Notice that before we observe $R$, $P(C_{RR}) = 1/3$. So,

\begin{align*}
P(C_{RR} | R) &= \frac{P(R | C_{RR}) P(C_{RR})}{P(R | C_{RR}) P(C_{RR}) + P(R | C_{RW}) P(C_{RR}) + P(R | C_{WW}) P(C_{RR})} \\
&= \frac{1/3 \times  1}{1/3 (1 + 1/2 + 0)} = 2/3
\end{align*}
